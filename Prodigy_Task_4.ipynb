```python
def plot_analysis(df, column_name):
    sent_analysis = df[column_name].value_counts(normalize=True)
    sentiments = sent_analysis.index.values
    vals = sent_analysis.values * 100  # Multiply the values by 100 for percentage
    plt.figure(figsize=(15, 15))
    k = sns.barplot(x=sentiments, y=vals)  # Specify x and y parameters
    for i in k.patches:
        k.annotate(format(i.get_height(), '.2f'),
                   (i.get_x() + i.get_width() / 2., i.get_height()),
                   ha='center', va='center',
                   xytext=(0, 9),
                   textcoords='offset points')
    plt.xlabel('sentiment')
    plt.ylabel('percentage')
    plt.xticks(rotation=90)
    plt.tight_layout()

# Call the function with the correct column name
plot_analysis(train_ds, 'Positive')
```

```python
plot_analysis(train_ds, 'Borderlands')
```
